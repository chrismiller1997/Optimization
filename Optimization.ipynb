```python
# call libraries
import numpy as np
from scipy.optimize import minimize
```

```python
# create the objective funtion
def objective(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    return x1 * x4 * (x1 + x2 + x3) + x3

# create the first constraint
def constraint1(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    # the right hand side of the inequality must be equal to zero
    return x1*x2*x3*x4 - 25.0

# create the second constraint
def constraint2(x):
    sum_sq = 40
    
    for i in range(4):
        sum_sq -= x[i]**2
    # again, the constraint equation must equal zero
    return sum_sq
```

```python
# test the initial condition
x0 = [1, 5, 5, 1]
print(objective(x0))
```

```python
# format the bounds and constraints 
b = (1.0, 5.0)
bounds = (b, b, b, b)
c1 = {'type': 'ineq', 'fun': constraint1}
c2 = {'type': 'eq', 'fun': constraint2}
constraints = [c1, c2]

# solve the minimization problem
solution = minimize(objective, x0, method='SLSQP', bounds=bounds, constraints=constraints)
print(solution)
```

The minimization is complete with a minimum value of 17.01401724563517 and variables x1 = 1.0, x2 = 4.7429961, x3 = 3.82115462, and x4 = 1.37940765.
